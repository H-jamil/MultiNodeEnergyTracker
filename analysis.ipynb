```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('test.csv')

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

duration_cols = ['DLDuration', 'DCopyDuration', 'FPDuration', 'BPDuration']
df[duration_cols].boxplot(ax=ax1)
ax1.set_title('Distribution of Durations')
ax1.set_ylabel('Duration (s)')

metrics = {
    'DL': ['DLCpuCore', 'DlCpuTotal', 'DLGpu'],
    'DCopy': ['DCopyCpuCore', 'DCopyCpuTotal', 'DCopyGpu'],
    'FP': ['FPCpuCore', 'FPCpuTotal', 'FPGpu'],
    'BP': ['BPCpuCore', 'BPCpuTotal', 'BPGpu']
}

means = {key: df[cols].mean() for key, cols in metrics.items()}

labels = list(metrics.keys())
cpu_core = [means[key][0] for key in labels]
cpu_total = [means[key][1] for key in labels]
gpu = [means[key][2] for key in labels]

x = np.arange(len(labels))
width = 0.35

ax2.bar(x, cpu_core, width, label='CPU Core')
ax2.bar(x, cpu_total, width, bottom=cpu_core, label='CPU Total')
ax2.bar(x, gpu, width, bottom=np.array(cpu_core) + np.array(cpu_total), label='GPU')

ax2.set_title('Average Energy Consumption by Component')
ax2.set_ylabel('Energy')
ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax2.legend()

plt.tight_layout()
plt.show()
```
